```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
```

### Lê o dataset no local que ele se emcontra e no formato que ele se encontra


```python
titanic = pd.read_csv("titanic/titanic.csv")
```

# Transformar para DataFrame a variável com o dataset

```python
titan = pd.DataFrame(titanic)
```



# Preencher valores NaN na coluna 'Survived' com 0, pois a IA não lê valores diferêntes de 0 e 1

```python
titan["Survived"].fillna(0, inplace=True)
```


# Criar variáveis dummy para a coluna 'Sex'

```python
titan = pd.get_dummies(titan, columns=["Sex"], drop_first=True)
```

# Preencher valores NaN nas colunas especificadas com 0. Novamente, a IA não pode receber valores NaN, pois não serão interpletados por ela.

```python
titan["Age"].fillna(0, inplace=True)
titan["PassengerId"].fillna(0, inplace=True)
```

 Separar variáveis independentes X da variável dependente y
 . As variáveis x e y são as colunas ajustadas para que a IA possa manipular.

```python
X = titan[['Pclass', 'Sex_male', 'Age']]
y = titan["Survived"]
```

# Dividir em treino e teste (80% treino, 20% teste).

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```


# Treinando o modelo

```python
regressor = LogisticRegression()
regressor.fit(X_train, y_train)
```


#previsões

```python
previsao = regressor.predict(X_test)
```

# Exibindo as previsões

```python
print(f"E a predição dos dados é {previsao}")
```
